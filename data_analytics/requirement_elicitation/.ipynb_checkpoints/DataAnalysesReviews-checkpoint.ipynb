```python
import pandas as pd

pd.set_option('display.max_rows', 100)
df = pd.read_csv('dublinNextBus.csv')
```

```python
df.head(50)
```

```python
# counts the most common words in the review column

pd.Series(' '.join(df['review']).lower().split()).value_counts()[:30]
```

```python
# counts the most frequently repeated phrases

from nltk import ngrams
vals = [y for x in df['review'] for y in x.split()]

n = [3,4,5]
a = pd.Series([y for x in n for y in ngrams(vals, x)]).value_counts()
print (a)
```

```python
# count the most frequently repeated phrases - takes off tuple

n = [3,4,5]
a = pd.Series([' '.join(y) for x in n for y in ngrams(vals, x)]).value_counts()
print (a)
```

```python
# counter
from nltk import ngrams
from collections import Counter

vals = [y for x in df['review'] for y in x.split()]
c = Counter([' '.join(y) for x in [3,4,5] for y in ngrams(vals, x)])

df1 = pd.DataFrame({'ngrams': list(c.keys()),
                   'count': list(c.values())})
print (df1)
```

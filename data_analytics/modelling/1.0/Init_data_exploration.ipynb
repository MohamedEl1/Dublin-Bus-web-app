```python
import pandas as pd
import os, sys
import numpy as np
from tqdm import tqdm_notebook
```

```python
# read leavetimes 2018 .txt file into csv leave16 (just first 100)
df = pd.read_csv('D:\data\\rt_leavetimes_DB_2018.txt', delimiter=';', chunksize=1000)


# just put 10 in a dataframe for the moment...
dfchunk=df.get_chunk(10)
dfchunk
```

```python
dfchunk.dtypes
```

```python
chunk_size = 1000000
suppressedIsOneCount = note_not_missing = pass_not_missing = passin_not_missing = passout_not_missing = total_dup_rows = db_as_source = 0
#use tqdm
# load the big file in smaller chunks
for count, dfchunk in tqdm_notebook(enumerate(pd.read_csv('D:\data\\rt_leavetimes_DB_2018.txt', delimiter=';', chunksize=chunk_size))):
    null_dict = {}
    print("--------------------")
    print("Chunk number: ", count+1)
    na = dfchunk.isna().sum()
    print(na)
    note_not_missing += chunk_size - na['NOTE']
    pass_not_missing += chunk_size - na['PASSENGERS']
    passin_not_missing += chunk_size - na['PASSENGERSIN']
    passout_not_missing += chunk_size - na['PASSENGERSOUT']
    # check for duplicate rows
#     print('Duplicate rows:', dfchunk.duplicated().sum())
    total_dup_rows += dfchunk.duplicated().sum()
    print("FINISHED A CHUNK")
    print()
    if any(dfchunk['DATASOURCE']!="DB"):
        db_as_source+= 1
    if any(dfchunk['SUPPRESSED']==1):
        suppressedIsOneCount+= 1
#     if count == 2:
#         break
print()
print("Finished")
print("Total no. of suppressed = 1 values: ", suppressedIsOneCount)
print("Total no. of Note not missing value: ", note_not_missing)
print("Total no. of Passengers not missing value: ", pass_not_missing)
print("Total no. of PassengersIn not missing value: ", passin_not_missing)
print("Total no. of PassengersOut not missing value: ", passout_not_missing)
print("Total no. of duplciated rows: ", total_dup_rows)
print("Total no. of data sources not DB: ", db_as_source)

    

```

## Initial findings
* Passengers, and passengers in and out are essentially all missing values and so are recommended to be dropped.
* Note is always missing and so will be dropped.
* Although there are some entries in 'justificationid', as the feature records 'fault codes' to do with why a journey was suppressed - this is not relevant to developing our model, so we will be dropping this column.
* Similarly with VehicleID
* lastupdate is not relevant to our model either, as it does not relate to what day a journey took place, only when the data was updated. So we will drop this column completely also.
* Similarly distance is dropped
* The feature datasource has a cardinality of one: ie, it only has one value, DB, for the entire dataset. Therefore we will drop this feature.

```python
chunk_size = 1000000
# load the big file in smaller chunks
output_path = 'D:\data\\clean_leavetimes.csv'
for count, dfchunk in tqdm_notebook(enumerate(pd.read_csv('D:\data\\rt_leavetimes_DB_2018.txt', delimiter=';', chunksize=chunk_size))):
    dfchunk.drop(columns=['DATASOURCE', 'PASSENGERS', 'PASSENGERSIN', 'PASSENGERSOUT', 'DISTANCE', 'SUPPRESSED', 'VEHICLEID', 'JUSTIFICATIONID', 'LASTUPDATE', 'NOTE'], inplace=True)
    dfchunk.to_csv(output_path, mode='a', header=not os.path.exists(output_path), index=False)
```

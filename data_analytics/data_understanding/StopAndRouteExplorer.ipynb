```python
import pandas as pd
import json
```

```python
df = pd.read_csv('rt_leavetimes_DB_2018.txt', delimiter=';', chunksize=10**6)
```

```python
stops={}
for chunk in df:
    for stop_point in chunk["STOPPOINTID"].unique():
        stops[stop_point]="n/a"
```

The stops are more than in the stops DB. examine which stops do not show up in the DB / vice versa:

```python
stops_df=pd.read_csv("stops_cleaned.txt")
```

```python
stops_df.head()
```

```python
stops_DB={}
for row in stops_df["stop_id"]:
    stops_DB[row]="n/a"
len(stops_DB)
```

```python
old_stops=[] #in training data but not in GTFS
new_stops=[] #in GTFS data but not in training data
for row in stops:
    if row not in stops_DB:
        old_stops.append(int(row))
for row in stops_DB:
    if row not in stops:
        new_stops.append(int(row))
print(f"Stop IDs in training data that are not in the DB: {len(old_stops)}")
print(f"Stop IDs in DB that are not in the training data: {len(new_stops)}")
```

```python
stop_json={
    "old_stops":old_stops,
    "new_stops":new_stops
}
```

```python
trips = pd.read_csv('rt_trips_DB_2018.txt',sep=";")
```

```python
trips.head()
```

```python
routes={}
for row in trips.ROUTEID:
    split=row.split("_")
    if split[0] not in routes:
        routes[split[0]]={split[1]:1}
    elif split[1] not in routes[split[0]]:
        routes[split[0]][split[1]]=1
    else:
        routes[split[0]][split[1]]+=1
```

```python
routes
```

```python
with open('route_combinations.json', 'w') as out:
    json.dump(routes, out)
```
